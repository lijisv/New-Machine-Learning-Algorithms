```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
days=[2,3,4,5,6,7]
temp=[10,100,30,90,50,60]
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.plot(days,temp,color='blue',linestyle='dashdot',marker='o',linewidth=2,alpha=0.6)# alpha is used for graph intensity
plt.show()
```

# legend grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[20,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather info')

plt.plot(days,max_t,label="max",linestyle='dashed')
plt.plot(days,min_t,label="min",linestyle='dashdot')
plt.plot(days,avg_t,label="avg",linestyle=':')
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
plt.show()
```
