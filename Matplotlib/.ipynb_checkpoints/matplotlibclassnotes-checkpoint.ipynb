```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("New Agenda")
plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label="profit")
plt.legend()
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("New Agenda")
plt.bar(company,profit)
plt.show()
```

```python
len(company)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("New Agenda")
plt.bar(xpos-0.2,revenue,width=-0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=+0.4,label="Profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("New Agenda")
##lt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
plt.show()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(15)
data = np.random.normal(100, 20, 200)
```

```python
data
```

```python
plt.boxplot(data)
fig = plt.figure(figsize =(10, 5))
plt.show()
```

```python
plt.hist(data,rwidth=0.8)
```

```python
np.random.seed(10)
fig = plt.figure(figsize =(10, 7))
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]


plt.boxplot(data,vert=True,patch_artist=True)
plt.legend()
plt.show()
```

```python
np.random.seed(10)

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]

fig = plt.figure(figsize =(10, 10))
plt.boxplot(data,vert=False,patch_artist=True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.axis("equal")

plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct="%0.1f%%",shadow=True,explode=[0.1,0,0.1,0.2,0],startangle=45)
plt.legend()
##plt.show()
plt.savefig("piechart_new.jpg")
```
